## Splitting the data into Independent variable (X) and dependent variable (y)

```python
X = diabetes[["bmi"]]
y = diabetes[["dis_prog"]]
```

```python
X
```

```python
y
```

## Splitting the data into train and test 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

It is clearly observalble that the data has been divided into 80% (for training data) and 20% for (Test data) 

```python
print(f" the shape of X_train data is {X_train.shape}")
print(f" the shape of y_train data is {X_test.shape}")
print(f" the shape of y_train data is {y_train.shape}")
print(f" the shape of y_test is {y_test.shape}")
```

## Visualizing the training and test data

```python
plt.scatter(X_train,y_train,color = 'blue',label = "Training_data")
plt.xlabel('Body Mass Index')
plt.ylabel('dis_prog')
```

```python
plt.scatter(X_test,y_test,color = "green",label = "Test data")
plt.xlabel('Body Mass Index')
plt.ylabel('dis_prog')
```

## Fitting the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print(f"The model coefficient for the model is {model.coef_}")
print(f"The intercept of the model is {model.intercept_}")
```

<b>Observation on coefficient and intercept</b>:-
    
the positive coefficient suggests that as the value of independent variable increases the value of dependent features increases
shows a positive correlation. 

The value of intercept as 151.87 says that the best fit line of the model toches 151.81 on y-axis

## Making predictions

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

## Calculating the loss on training and test data using mean squared error

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
train_loss = mean_squared_error(y_train_pred,y_train)
test_loss = mean_squared_error(y_test_pred,y_test)
```

#### Train Loss

## Plotting the regression (prediction) line vs the actual data points 

```python
plt.scatter(X_train,y_train,color = 'blue',label = "Training_data")
plt.plot(X_train,y_train_pred,color = "green",linewidth=3,label = "Prediction Line")
plt.title("Training data: BMI Vs dis_prog")
plt.xlabel('Body Mass Index')
plt.ylabel('dis_prog')
```

```python
plt.scatter(X_test,y_test,color = 'red',label = "Training_data")
plt.plot(X_test,y_test_pred,color = "yellow",linewidth=3,label = "Prediction Line")
plt.title("Test data: BMI Vs dis_prog")
plt.xlabel('Body Mass Index')
plt.ylabel('dis_prog')
```

## Performing cross validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X,y,cv = 5)
```

#### As r2 is the default metric for regression accuarcy calculation. Calculated r2, Mean Absolute Error (MAE). And Mean Squared Error (MSE)

```python
print("Check for Regression metrics on the train set")
print(f"R2 Score:{r2_score(y_test,y_test_pred)}")
print(f"MAE:{mean_absolute_error(y_test,y_test_pred)}")
print(f"MSE:{mean_squared_error(y_test,y_test_pred)}")
```

```python
print("Check for Regression metrics on the train set")
print(f"R2 Score:{r2_score(y_train,y_train_pred)}")
print(f"MAE:{mean_absolute_error(y_train,y_train_pred)}")
print(f"MSE:{mean_squared_error(y_train,y_train_pred)}")
```

### Observations and findings 

My Observations - This particular regression model is built using bmi as an independent feature and dis_prog (the feature which is predicted)as a dependent feature. The reason for a very high MSE, MAE and low accuracy is as :- As there is only one predictor (independent feature)it makes the model incompetent to make prediction as there are no other estimators.
