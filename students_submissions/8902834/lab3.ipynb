```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python

diabetes_bmi = diabetes_X[:,2].reshape((-1,1))
X_train,X_test,y_train,y_test = train_test_split(diabetes_bmi, diabetes_y, test_size=0.2, random_state= 0)
```

<h3><U><B>We built a Linear Regression Model that fits our Training Datasets:</B></U></h3>

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

<h3><B><U>The Model's Parameter - Slope(m) and Intercept(c):</U></B></h3>

```python
m = float(model.coef_)
c = model.intercept_
y_pred = model.predict(X_test)
print("The slope of the model is m =",m)
print("The intercept of the model is c =",c)
```

<h3><u><b>The loss on the Training and Testing Datasets:</B></U></h3>

```python
y_train_pred = model.predict(X_train)
train_loss = np.mean((y_train - y_train_pred) ** 2)
test_loss = np.mean((y_test - y_pred) ** 2)
print("The loss on the training dataset is", train_loss)
print("The loss on the testing dataset is", test_loss)
```

<h3><u><b>A graph representing the Training Dataset:</h3></u></b>

```python
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Training Data')
plt.legend()
plt.show()

```

<h3><u><b>A graph representing the Testing Datatset:</h3></u></b>

```python
plt.scatter(X_test, y_test, color='blue', label='Testing Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Testing Data')
plt.legend()
plt.show()
```

<h3><u><b>A graph respresenting the Linear Regression - Model:</h3></u></b>

```python
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Model')
plt.legend()
plt.show()
```

<h3><u><b>A graph respresenting the Linear Regression - Model, Training and Testing Dataset:</h3></u></b>

```python


# Plot the test data and the linear regression model
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.scatter(X_test, y_test, color='blue', label='Test Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Model, Training and Testing')
plt.legend()
plt.show()
```

<h3><u><b>Takeaways:</h3></u></b>

1. The model's slope (m) indicates the rate of change in disease progression per unit change in BMI. A positive value suggests that higher BMI is associated with increased disease progression.
2. The intercept (c) represents the estimated disease progression when BMI is zero. However, in the context of BMI, an intercept of zero is not meaningful since BMI cannot be zero in reality.
3. The loss on the test dataset provides an indication of how well the model generalizes to unseen data. A lower loss value indicates better performance.

To conclude, the univariate linear regression model on the diabetes dataset provides a reasonable fit, suggesting that BMI may be a useful predictor for disease progression. However, further analysis and evaluation of additional features could provide a more comprehensive understanding of the factors influencing disease progression in diabetes.
