Loading Data set

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

iris = load_iris(as_frame=True)
```

```python
iris.target_names
```

```python
iris.data
```

```python
iris.target_names
print(iris.DESCR)
```

```python
X = iris.data
y = iris.target_names[iris.target] == 'virginica'
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.75, random_state=42)
```

```python
LR = LogisticRegression(random_state=42)
LR.fit(X_train, y_train)
```

```python
y_test
```

```python
y_predict_test = LR.predict(X_test)
y_predict_test
```

In order to evaluate logistic regression we use confusion matrix.

```python
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, y_predict_test)
print(cnf_matrix)
```

There are no false positive and false negatives.

Cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(LR, X_train, y_train, cv=10, scoring='accuracy')
print(scores)
scores.mean()
```

Precision and recall

```python
from sklearn.metrics import precision_score
prec_score = precision_score(y_test, y_predict_test)
prec_score
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_predict_test)
recall
```

The model is giving 1 for precision and recall. And in cross validation using 10 split it gives an accuarcy mean of 95.45%.
