```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt  
```

```python
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X.shape
```

```python
diabetes_y.shape
```

```python
df_diabetes_X = pd.DataFrame(diabetes_X)
df_diabetes_X.columns = ['age','sex','body_mass_index','average_blood_pressure','total_serum_cholesterol','low_density_lipoproteins','high_density_lipoproteins','total_cholesterol','possibly_log_serum_triglycerides_level','blood_sugar_level']

df_diabetes_y = pd.DataFrame(diabetes_y)
df_diabetes_y.columns = ['disease_progression_after_year']
```

```python
plt.scatter(df_diabetes_X['body_mass_index'], df_diabetes_y['disease_progression_after_year'])
plt.xlabel("Body Mass Index(bmi)")
plt.ylabel("Disease Progression")
plt.title("Disease progression after one year based on Body Mass Index ")
plt.show()
```

```python
diabetes = pd.DataFrame(diabetes_X,columns = ['age','sex','body_mass_index','average_blood_pressure','total_serum_cholesterol','low_density_lipoproteins','high_density_lipoproteins','total_cholesterol','possibly_log_serum_triglycerides_level','blood_sugar_level'])
diabetes['disease_progression_after_year'] = pd.DataFrame(diabetes_y)
```

```python
X = diabetes[['body_mass_index']]
y = diabetes[['disease_progression_after_year']]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()  
model.fit(X_train, y_train)
```

```python
print('Coefficients(Slope) :', model.coef_)
```

```python
print('Intercept is', model.intercept_[0])
```

```python
y_pred_arr = model.predict(X_test)
y_pred_test = pd.DataFrame(y_pred_arr)
```

## Fitting the model with test data


```python
plt.scatter(X_test['body_mass_index'], y_test,  color='red')
plt.plot(X_test['body_mass_index'], y_pred_test, color='blue', linewidth=1)
plt.xlabel("Body Mass Index(bmi)")
plt.ylabel("Disease Progression")
plt.title("Disease progression after one year based on Body Mass Index (Test Data) ")
plt.show()
```

```python
print('Mean squared error for test data : %.2f' % metrics.mean_squared_error(y_test, y_pred_test))
```

```python
y_pred = model.predict(X_train)
y_pred_train = pd.DataFrame(y_pred)
```

## Fitting the model with train data

```python
plt.scatter(X_train['body_mass_index'], y_train,  color='orange')
plt.plot(X_train['body_mass_index'], y_pred_train, color='green', linewidth=2)
plt.xlabel("Body Mass Index(bmi)")
plt.ylabel("Disease Progression")
plt.title("Disease progression after one year based on Body Mass Index (Train Data) ")
plt.show()
```

```python
print('Coefficients(Slope) :', model.coef_)
```

```python
print('Accuracy of model is', model.score(X,y))
```
